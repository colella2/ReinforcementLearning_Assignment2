Please use the following equations. 

s' = a + e'

s' = (s-c) + e'

e' must be known after the decision "a" is made. 

```python
import pandas as pd
import numpy as np
```

```python
Pt_e = np.array(
            [[0.8, 0.1, 0.1],
               [0.01, 0.98, 0.01],
               [0.1, 0.1, 0.8]])
```

```python
Pt_e
```

```python
col_name = []
for s in range(5):
    for e in range(3):
        col_name.append((s,e))
```

```python
col_name #(s, e)
```

```python
Pt_se = np.empty((15,15))
Pt_se[:] = np.nan
Pt_se_df = pd.DataFrame(Pt_se, columns=col_name, index=col_name)
```

```python
Pt_se_df
```

```python
# Given that with e cannot be larger than s as s = s(t=t-1) + e, therefore, probability on transitioning from (s,e) where s < e should be 0.
#looping through (s, e) and (s', e') to drop off rows and columns with 0 probability

for (s,e) in col_name:
    if s < e:
        Pt_se_df.drop([(s,e)],inplace=True)
for(s_,e_) in col_name: 
    if s_ < e_:
        Pt_se_df.drop(columns=[(s_,e_)], inplace=True)      

col_name = list(Pt_se_df)
```

```python
Pt_se_df
```

```python
Pt_se_df_0 = Pt_se_df.copy()
Pt_se_df_1 = Pt_se_df.copy()
Pt_se_df_2 = Pt_se_df.copy()
```

```python
# For a = 0

# Given that with a = 0, s' must be equal to e', therefore, probability on transitioning to (s',e') where s' != e' should be 0.
#looping through (s', e')
for (s_,e_) in col_name:
    if s_ != e_:
        Pt_se_df_0.loc[:, (s_,e_)] = 0

# Transitioning from (s,e) to (s',e') should now only be dependent on transition probability matrix (e,e')
for (s,e) in col_name:
    for(s_,e_) in col_name:
        if Pt_se_df_0.at[(s,e), (s_,e_)] != 0:
            Pt_se_df_0.at[(s,e), (s_,e_)] = Pt_e[e,e_]
```

```python
Pt_se_df_0
```

```python
# For a = 1
        
# Given that with a = 1, s' must be equal to e', therefore, probability on transitioning to (s',e') where s' != e' + 1 should be 0.
#looping through (s', e')
for (s_,e_) in col_name:
    if s_ != e_ + 1:
        Pt_se_df_1.loc[:, (s_,e_)] = 0
        
# Transitioning from (s,e) to (s',e') should now only be dependent on transition probability matrix (e,e')
for (s,e) in col_name:
    for(s_,e_) in col_name:
        if Pt_se_df_1.at[(s,e), (s_,e_)] != 0:
            Pt_se_df_1.at[(s,e), (s_,e_)] = Pt_e[e,e_]
```

```python
Pt_se_df_1
```

```python
# For a = 2
        
# Given that with a = 1, s' must be equal to e', therefore, probability on transitioning to (s',e') where s' != e' + 2 should be 0.
#looping through (s', e')
for (s_,e_) in col_name:
    if s_ != e_ + 2:
        Pt_se_df_2.loc[:, (s_,e_)] = 0
        
# Transitioning from (s,e) to (s',e') should now only be dependent on transition probability matrix (e,e')
for (s,e) in col_name:
    for(s_,e_) in col_name:
        if Pt_se_df_2.at[(s,e), (s_,e_)] != 0:
            Pt_se_df_2.at[(s,e), (s_,e_)] = Pt_e[e,e_]
```

```python
Pt_se_df_2
```

```python
def reward(c):
    if c >= 0:
        return np.log(c + 1)
    else:
        return np.NINF
```

```python
reward_row = []
for a in range(3):
    for (s,e) in col_name:
        reward_row.append((s,e,a))
```

```python
reward_row
```

```python
R_sea = np.empty((len(reward_row),len(col_name)))
R_sea[:] = np.nan
R_sea_df = pd.DataFrame(R_sea, columns=col_name, index=reward_row)
```

```python
R_sea_df
```

```python
#reward is dependent on availability of ice cream, s, at the initial state
for (s,e,a) in reward_row:
    c = s - a
    R_sea_df.loc[(s,e,a),:] = reward(c)
```

```python
R_sea_df
```

```python
R_se_df_0 = Pt_se_df.copy()
R_se_df_1 = Pt_se_df.copy()
R_se_df_2 = Pt_se_df.copy()
```

```python
#reward matrix for a = 0 i.e. eating all the ice cream (s = c)
a = 0
for (s,e) in col_name:
    c = s - a
    R_se_df_0.loc[(s,e),:] = reward(c)
      
a = 1
for (s,e) in col_name:
    c = s - a
    R_se_df_1.loc[(s,e),:] = reward(c)

a = 2
for (s,e) in col_name:
    c = s - a
    R_se_df_2.loc[(s,e),:] = reward(c)
```

```python
R_se_df_0
```

```python
P = np.array([Pt_se_df_0, Pt_se_df_1, Pt_se_df_2],dtype=float)

K,N,M = P.shape
```

```python
RR = np.array([R_se_df_0, R_se_df_1, R_se_df_2],dtype=float)
```

```python
RR
```

```python
RR.shape
```

```python
P[0,:,:]
```

```python
K
```

```python
q = np.zeros((K,N,1),dtype=np.float64) 
for i in range(0,N):
    for k in range(0,K):
        for j in range(0,M):
            q[k,i,0] = q[k,i,0] + RR[k,i,j]*P[k,i,j]
```

```python
T=600
v = np.zeros((N,T),dtype=np.float64)
d = np.zeros((N,T),dtype=np.float64)
for n in range(1,T):
    for i in range(0,N):
        rhs = np.zeros((1,K),dtype=np.float64)
        for k in range(0,K):
            rhs[0][k] = q[k][i][0] + np.matmul(P[k,i,:],v[:,n-1]) 
        v[i,n] = max(rhs[0])
        d[i,n] = np.argmax(rhs[0])
v = v.T
d = d.T
```
