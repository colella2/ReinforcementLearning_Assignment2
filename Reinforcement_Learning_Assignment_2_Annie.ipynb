```
#(15x15 matrix)
index=[]
for s in range(5):
  for e in range(3):
    index.append((s,e)) 

zeroes=np.zeros((15,15))
trans_mat = pd.DataFrame(zeroes,index=index,columns=index)
trans_mat
```

Action = 0

When a = 0, s' = e', so all else elements should be zero

```
trans_mat_0=trans_mat.copy()
for (s,e) in index:
  for (s_,e_) in index:
    if (s_==e_) & (s>=e):
     trans_mat_0.at[(s,e),(s_,e_)]=P_mat[e,e_]
trans_mat_0
```

Conditions:
